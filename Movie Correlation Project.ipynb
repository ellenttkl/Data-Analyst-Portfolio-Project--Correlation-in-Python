```python
# import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
import matplotlib

plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8)#Adjust the configuration of the plots we will create


#read in the data

df = pd.read_csv(r'C:\Users\16032\Desktop\movies.csv')
```

```python
# data detail information

df.head()
```

```python
# any missing data?

for col in df.columns:
   pct_missing = np.mean(df[col].isnull())
   print('{}-{}'.format(col,pct_missing))
```

```python
# data type of columns
df.dtypes
```

```python
#change data type of columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')

# or

# df = df.astype({"budget":'object',"gross":'object'})
```

```python
# create correct year column

df['yearcorrect'] = df['released'].astype('string').str[:4]
```

```python
# sort the data by 'gross' column

df = df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
# display all rows of the data

pd.set_option('display.max_rows',None)
```

```python
# drop any duplicates

#df['company'].drop_duplicates().sort_values(ascending=False)
#df['company'] = df['company'].drop_duplicates()
df.drop_duplications()
```

```python
# scatter plot with budget vs gross
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
# Plot Budget vs Gross using Seaborn

sns.regplot(x=df['budget'],y=df['gross'],scatter_kws={'color':'red'},line_kws={'color':'green'})
```

```python
# let's looking at correlation

df.corr(method='pearson')#pearson,kendall,spearman
```

```python
# High correlation between budget and gross

correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
